```python
import numpy as np
import pandas as pd
```

```python
dbpath = 'mydb.sqlite'
conn = sqlite3.connect(dbpath)
cur = conn.cursor()
```

```python
cur.executescript('DROP TABLE IF EXISTS word_counts;')
```

```python
cur.executescript("""
CREATE TABLE stories(
    story_id INTEGER PRIMARY KEY UNIQUE,
    words TEXT
    );
""")
```

```python
cur.executescript("""
CREATE TABLE word_counts(
    word_count_id INTEGER PRIMARY KEY UNIQUE,
    word TEXT,
    count INTEGER,
    story_id INTEGER,
    FOREIGN KEY (story_id)
        REFERENCES stories (story_id)
            ON DELETE CASCADE
            ON UPDATE NO ACTION
    );
""")
```

```python
story_file = open("story-4.txt",encoding='utf-8')
word_counter = {}
counter = 0

for line in story_file:
    line_list = line.split()
    for word in line_list:
        real_word = word.lower()
        new_word = ''.join(char for char in real_word if char.isalnum())
        if new_word not in word_counter.keys():
            word_counter[new_word] = 1
        else:
            add_value = word_counter.get(new_word)
            add_value += 1
            word_counter[new_word] = add_value
print(word_counter)

    
story_file.close()
```

```python
story_list = ["story-1.txt","story-2.txt","story-3.txt","story-4.txt"]
story_count = 1
for story in story_list:
    story_file_loop = open(story,encoding='utf-8')
    story_text = story_file_loop.read()
    cur.execute("""INSERT INTO stories(story_id,words) VALUES (?,?);""",(story_count,story_text))
    conn.commit()
    story_count += 1
    story_file_loop.close()
```
